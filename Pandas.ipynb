```python
adf=pand.DataFrame({"x1":["m","n","l"],
                   "x2":[4,5,6]})
```

```python
bdf=pand.DataFrame({"x1":["m","n","p"],
                    "x3":["T","F","T"]})
```

```python
bdf
```

```python
cdf=pand.DataFrame({"x1":["n","l","p"],
                    "x2":[5,6,7]})
```

```python
cdf
```

```python
pand.merge(adf,bdf,how='left',on='x1')
```

```python
pand.merge(adf,bdf,how='right',on='x1')
```

```python
pand.merge(adf,bdf,how='inner',on='x1')
```

```python
pand.merge(adf,bdf,how='outer',on='x1')
```

```python
adf[adf.x1.isin(bdf.x1)]
```

```python
adf[~adf.x1.isin(bdf.x1)]
```

```python
pand.merge(adf,cdf)
```

```python
pand.merge(adf,cdf,how='outer')
```

```python
pand.merge(adf,cdf,how='outer',indicator=True)
```

```python
pand.merge(adf,cdf,how='outer',indicator=True).query('_merge=="right_only"')
```
